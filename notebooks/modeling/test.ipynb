```python
from sklearn.model_selection import train_test_split
import pandas as pd
from path import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sqlalchemy import create_engine
import datetime
import numpy as np
```

```python
file_path = Path("../../data/zillow_ca_la.csv")

ca_df = pd.read_csv(file_path)
ca_df.head()
```

```python
#Declaring features and targets

y= ca_df["AvgSale_Price"]

X= ca_df["No_of_bedrooms"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
#instantized and fit data to model and predictions
model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_train)
```

```python
#Print the coeffiecient and intercept of the nodel

print(model.coef_)
print(model.intercept_)
```

```python
#Print regression line

plt.scatter(X,y)
plt.plot(X, y_pred, color="red")
plt.show()
```
